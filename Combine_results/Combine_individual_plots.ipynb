```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python

subject='SG11'
```

```python
riem=np.load('Individual_accuracy/Riemann_'+subject+'.npy')
csp=np.load('Individual_accuracy/CSP_'+subject+'.npy')
source_average=np.load('Individual_accuracy/Source_data_'+'SG16'+'_average.npy')[:,:84]
source=np.load('Individual_accuracy/Source_data_'+subject+'.npy')[:,:84]
```

```python
plt.style.use('ggplot')
```

```python

w_times=np.linspace(1,5,len(np.mean(riem,0)))
```

```python
aplha=0.6
matplotlib.rcParams.update({'font.size': 22})
plt.figure(figsize=(15,10))

plt.plot(w_times,np.mean(riem,0),label='Riemann',color='b', alpha=aplha)
plt.plot(w_times,np.mean(csp,0),label='CSP',color='purple',alpha=aplha)
plt.plot(w_times,np.mean(source_average,0),label='Source Average MRI',color='r',alpha=aplha)
plt.plot(w_times,np.mean(source,0),label='Source Individual MRI',color='g',alpha=aplha)



plt.yticks()
plt.xlabel('Accuracy')
plt.ylabel('Time')


plt.legend()

my_dpi=150

plt.title(subject)
plt.savefig('Figures/'+subject+'.png', format='png', dpi=my_dpi)
```

```python

subjects=['SG01','SG02','SG05','SG12','SG11','SG15','SG16']
```

```python
riems=[]
csss=[]
source_averages=[]
sources=[]
for subject in subjects:
    riems.append(np.load('Individual_accuracy/Riemann_'+subject+'.npy'))
    csss.append(np.load('Individual_accuracy/CSP_'+subject+'.npy'))
    source_averages.append(np.load('Individual_accuracy/Source_data_'+subject+'_average.npy')[:,:84])
    sources.append(np.load('Individual_accuracy/Source_data_'+subject+'.npy')[:,:84])
    
riems=np.array(riems)
csss=np.array(css)
source_averages=np.array(source_averages)
sources=np.array(sources)
```

```python
mean_riems=np.mean(riems,(0,1))
std_riems=0.8*np.mean(stats.sem(riems),0)

mean_csss=np.mean(csss,(0,1))
std_csss=0.8*np.mean(stats.sem(csss),0)


mean_source_averages=np.mean(source_averages,(0,1))
std_source_averages=0.8*np.mean(stats.sem(source_averages),0)

mean_sources=np.mean(sources,(0,1))
std_sources=0.8*np.mean(stats.sem(sources),0)
```

```python
plt.plot(np.mean(riems,(0,1)))
plt.plot(np.mean(csss,(0,1)))
plt.plot(np.mean(source_averages,(0,1)))
plt.plot(np.mean(sources,(0,1)))




aplha=0.8
matplotlib.rcParams.update({'font.size': 22})
plt.figure(figsize=(15,10))

plt.plot(w_times,mean_riems,label='Riemann',color='b', alpha=aplha)
plt.plot(w_times,mean_riems+std_riems,'b',alpha=0.3)
plt.plot(w_times,mean_riems-std_riems,'b',alpha=0.3)
#plt.plot(time_range,source_mean-source_std,'b',alpha=aplha)
plt.fill_between(w_times,mean_riems-std_riems,mean_riems+std_riems,color='b',alpha=0.2)




plt.plot(w_times,mean_csss,label='CSP',color='purple',alpha=aplha)
plt.plot(w_times,mean_csss+std_csss,color='purple',alpha=0.3)
plt.plot(w_times,mean_csss-std_csss,color='purple',alpha=0.3)
#plt.plot(time_range,source_mean-source_std,'b',alpha=aplha)
plt.fill_between(w_times,mean_csss-std_csss,mean_csss+std_csss,color='purple',alpha=0.2)




plt.plot(w_times,mean_source_averages,label='Source Average MRI',color='r',alpha=aplha)
plt.plot(w_times,mean_source_averages+std_source_averages,color='r',alpha=0.3)
plt.plot(w_times,mean_source_averages-std_source_averages,color='r',alpha=0.3)
#plt.plot(time_range,source_mean-source_std,'b',alpha=aplha)
plt.fill_between(w_times,mean_source_averages-std_source_averages,mean_source_averages+std_source_averages,color='r',alpha=0.2)


plt.plot(w_times,mean_sources,label='Source Individual MRI',color='g',alpha=aplha)
plt.plot(w_times,mean_sources+std_sources,color='g',alpha=0.3)
plt.plot(w_times,mean_sources-std_sources,color='g',alpha=0.3)
#plt.plot(time_range,source_mean-source_std,'b',alpha=aplha)
plt.fill_between(w_times,mean_sources-std_sources,mean_sources+std_sources,color='g',alpha=0.2)



scp_std=np.std(riems,(0,1))
riem_std=np.std(np.array([riem[i][:121] for i in range(len(riem))]),0)


plt.yticks()
plt.xlabel('Time')
plt.ylabel('Accuracy')


plt.legend()

my_dpi=150

plt.title('All subjects')
plt.savefig('Figures/All_subjects'+'.png', format='png', dpi=my_dpi)
```
