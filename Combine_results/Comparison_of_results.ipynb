```python
def extract_from_sources(df_source,task,subject):
    ddf=df_source[ (df_source['tasks']==task) & (df_source['subject']==subject)].groupby(['subject','tasks','clusters_num','atlasses','lambdas']).mean()
    acc_value=ddf.groupby(['subject','tasks']).max('acc').acc.values[0]
    
    
    cluster_v=ddf[ddf['acc']==acc_value].index[0][2]
    ROI_v=ddf[ddf['acc']==acc_value].index[0][3]
    lambda_v=ddf[ddf['acc']==acc_value].index[0][4]
    
    
    acc_values=df_source[ (df_source['tasks']==task) & (df_source['subject']==subject)
      & (df_source['clusters_num']==cluster_v)
       & (df_source['atlasses']==ROI_v)
       & (df_source['lambdas']==lambda_v)].groupby(['time']).mean()
    
    print(cluster_v)
    print(ROI_v)
    print(lambda_v)
    return acc_values['acc'].values
```

```python
def return_stats(kek, lol, riem):
    s,p=kruskal(kek,lol,riem)
    s=np.round(s,3)
    p=format(p, "5.2e")
    print('Kruskal result '+ str(p))
    print('Kruskal result '+ str(s))
    
    s,p=mannwhitneyu(kek,lol)
    p_1=p#format(p, "5.2e")
    s_1=np.round(s,3)
    #print(p)
    #print(s)
    s,p=mannwhitneyu(kek,riem)
    p_2=p#format(p, "5.2e")
    s_2=np.round(s,3)
    #print(p)
    #print(s)
    s_3,p=mannwhitneyu(lol,riem)
    p_3=p#format(p, "5.2e")
    s=np.round(s,3)
    #print(p)
    #print(s)
    a=multipletests(np.array([p_1,p_2,p_3]),method='fdr_bh')
    
    print('Source Space\tCSP\t'+str(s_1)+'\t'+str(format(a[1][0], "5.2e"))+'\t'+str(round(np.mean(kek)-np.mean(lol),3)))
    print('Source Space\tRiemann\t'+str(s_2)+'\t'+str(format(a[1][1], "5.2e"))+'\t'+str(round(np.mean(kek)-np.mean(riem),3)))
    print('CSP\tRiemann\t'+str(s_3)+'\t'+str(format(a[1][2], "5.2e"))+'\t'+str(round(np.mean(lol)-np.mean(riem),3)))
    
    
    return a[1]
```
